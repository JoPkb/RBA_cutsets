# Mise en valeur des fonctions m√©taboliques du foie sain:

### Utils

```python
def optimize_secretion(model, metabolite_id, exchanges=False, parcours=False):
    """
    Function used to optimize the secretion reaction of a given metabolite.
    Reverses the original exchange reaction if it is an intake.
    """
    with model as m:
        metabolite_id_x = metabolite_id[:-1]+"x"
        exchange_reaction = m.reactions.get_by_id(f"EX_{metabolite_id_x}")

        if len(exchange_reaction.reactants) == 0:
            coeff = [v for v in exchange_reaction.metabolites.values()][0]
            #revert reaction
            exchange_reaction.add_metabolites({metabolite_id_x : -(2*coeff)})
        

        m.objective = exchange_reaction.id
        sol = model.optimize()
        if exchanges:
            vu.print_exchanges(m, filter="all")
        elif parcours:
            print(exchange_reaction)
            vu.run_parcours(exchange_reaction.id,m)
    return sol
```

#### Fonction d√©finissant des r√©actions d'√©change de sortie pour les m√©tabolites qui n'en ont pas :

```python
def add_secretion_reaction(metabolite, model):
    reaction = cobra.core.reaction.Reaction(id=f"EXR_{metabolite.id}", name=f"{metabolite.id} secretion", subsystem="Exchange reactions",lower_bound=0.0,upper_bound=1000.0)
    reaction.add_metabolites({metabolite : -1.0})
    model.add_reactions([reaction])
```

```python
def add_rev_exchanges(model):
    #Copy the model:
    wmodel = model.copy()

    # get all exchange reactions:
    exchanges = [r for r in wmodel.boundary]

    # Keep only the reversible ones:

    reversible = [rev for rev in exchanges if rev.bounds == (-1000.0,1000.0)]

    # For each reversible reaction, create a reaction with opposite stoichiometry :

    for reac in reversible:
        if len(reac.reactants) == 0:
            metabolite = reac.products[0]
            rev_reac = cobra.core.reaction.Reaction(id=f"rev_EX_{metabolite.id}", name=f"{metabolite.id} secretion", subsystem="Exchange reactions", lower_bound=0.0, upper_bound=1000.0)
            rev_reac.add_metabolites({metabolite : -1})
            wmodel.add_reactions([rev_reac])
            print(f"Added {rev_reac.id} reaction")
        elif len(reac.products) == 0:
            print(reac)
    return wmodel
```
